### Objective of K-Means
The objective of k-means clustering is to partition a dataset into k distinct clusters. Each data point belongs to the cluster with the nearest mean, and the algorithm aims to minimize the within-cluster sum of squared distances. In simpler terms, k-means seeks to find groups in the data where the members within each group are more similar to each other than they are to those in other groups.

### How it Works
<img src="../images/k_means_example.png" width="550" />

The k-means algorithm follows theses steps:
1. **Initialization**: Randomly select k data poitns as initial centroids
2. **Assignment**: Assign each data point to the nearest cenroid, forming k clusters
3. **Update**: Recalcaulte the centroids by taking the mean of all data poitns within each cluster
4. **Repeat**: Repeat steps 2-3 until convergence (stationary centroids)

### Choosing k
One significant downside to k-means clustering is that you have to specify the number of clusters with k. Some methods for choosing k include:
- **Elbow Method**: This technique involves plotting the variance explained against the number of clusters. The key is to identify the "elbow" point on the graph, where the rate of variance decrease slows down.
- **Silhouette Analysis**: This approach evaluates how well-separated the clusters are. The silhouette value measures the similarity of a point to its own cluster compared to other clusters. A higher silhouette score indicates well-defined and distinct clusters.

## Demo

```python
# Imports
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Generate dummy data
np.random.seed(0)
cov = np.array([[1, 0], [0, 1]])
mean1 = np.array([2, 2])
cluster1 = np.random.multivariate_normal(mean1, cov, 20)
mean2 = np.array([8, 2])
cluster2 = np.random.multivariate_normal(mean2, cov, 20)
mean3 = np.array([4, 8])
cluster3 = np.random.multivariate_normal(mean3, cov, 20)
data = np.concatenate([cluster1, cluster2, cluster3])

# Plot the generated data
sns.set_theme()
sns.scatterplot(x=data[:, 0], y=data[:, 1], s=30)
plt.title("Raw Data")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.show()
```

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
kmeans.fit(data)

# Get cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Plot the data with cluster centers
sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=labels, s=20, palette=sns.color_palette('tab10', n_colors=3))
sns.scatterplot(x=centers[:, 0], y=centers[:, 1], color='red', marker='X', s=100, label='Centroids')
plt.title("KMeans Clustering")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
plt.show()
```
