### Demo

```python
# Imports
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Generate dummy data
np.random.seed(0)
cov = np.array([[1, 0], [0, 1]])
mean1 = np.array([2, 2])
cluster1 = np.random.multivariate_normal(mean1, cov, 20)
mean2 = np.array([8, 2])
cluster2 = np.random.multivariate_normal(mean2, cov, 20)
mean3 = np.array([4, 8])
cluster3 = np.random.multivariate_normal(mean3, cov, 20)
data = np.concatenate([cluster1, cluster2, cluster3])

# Plot the generated data
sns.set_theme()
sns.scatterplot(x=data[:, 0], y=data[:, 1], s=30)
plt.title("Raw Data")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.show()
```

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(data)

# Get cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Plot the data with cluster centers
sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=labels, s=20, palette=sns.color_palette('tab10'))
sns.scatterplot(x=centers[:, 0], y=centers[:, 1], color='red', marker='X', s=100, label='Centroids')
plt.title("KMeans Clustering")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
plt.show()
```
